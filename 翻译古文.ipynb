```python
import os
import jieba
from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
source_root = 'Classical-Modern/source'
target_root = 'Classical-Modern/target'
all_parallel = []
for f in os.listdir(source_root):
    print(f)
    source_file = os.path.join(source_root, f)
    target_file = os.path.join(target_root, f + '翻译')
    source = open(source_file).read().split('\n')
    target = open(target_file).read().split('\n')
    assert len(source) == len(target)
    parallel = list(zip(source, target))
    all_parallel += parallel
```

```python
print(len(all_parallel))
```

```python
def strip(x):
    x = x.strip()
    while x.endswith('。'):
        x = x[:-1]
    x = x.strip()
    return x
```

```python
all_parallel = [(strip(x[0]), strip(x[1])) for x in all_parallel]
all_parallel = [x for x in all_parallel if x[0] and x[1]]
```

```python
print(len(all_parallel))
```

```python
src = [' '.join(x[0]) for x in tqdm(all_parallel)]
tgt = [' '.join(jieba.lcut(x[1])) for x in tqdm(all_parallel)]
```

```python
print(src[10])
print(tgt[10])
```

```python
src_train, src_val, tgt_train, tgt_val = train_test_split(
    src, tgt, test_size=10000, random_state=227)
```

```python
# 这里交换一下src和tgt哦
# 普通话 -> 古文
with open('src-train.txt', 'w') as fp:
    fp.write('\n'.join(tgt_train))
with open('tgt-train.txt', 'w') as fp:
    fp.write('\n'.join(src_train))
with open('src-val.txt', 'w') as fp:
    fp.write('\n'.join(tgt_val))
with open('tgt-val.txt', 'w') as fp:
    fp.write('\n'.join(src_val))
```

```python
# https://opennmt.net/OpenNMT-py/options/train.html
# https://github.com/OpenNMT/OpenNMT-py/blob/master/docs/source/FAQ.md#how-do-i-use-the-transformer-model-do-you-support-multi-gpu
yml = f'''# train.yaml
## Where the samples will be written
save_data: {model_root}
## Where the vocab(s) will be written
src_vocab: {model_root}/src.vocab
tgt_vocab: {model_root}/src.vocab
src_vocab_size: 200000
tgt_vocab_size: 200000
share_vocab: true

# Batching
queue_size: 100
bucket_size: 2048
world_size: 1
gpu_ranks: [0]

batch_size: 32
valid_batch_size: 16

# Corpus opts:
data:
    corpus_1:
        path_src: src-train.txt
        path_tgt: tgt-train.txt
    valid:
        path_src: src-val.txt
        path_tgt: tgt-val.txt

# Where to save the checkpoints
save_model: {model_root}/model
save_checkpoint_steps: 10000
train_steps: 1000000
valid_steps: 10000
'''
with open('train.yaml', 'w') as fp:
    fp.write(yml)
```
